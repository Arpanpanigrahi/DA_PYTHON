# Correlation with HeatMap
1. A correlation will give you an inter relationship between two classes. 
2. It is applicable only on numerical values like int, float.
3. Value will be range from -1 to +1.

```python
df2 = df[['total_bill', 'tip', 'size']]
df2
# df.corr()
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr())
```

# Univariant Analysis

```python
#kde - kernal density information
sns.distplot(df['size'],kde=False)
```

```python
sns.distplot(df['tip'])
```

```python
sns.displot(df['tip'], kde=False, bins=10)
```

# Bivariant Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# SEABORN-1
1. Seaborn is a library used for making statistical graph (mean, mode, median) in Python.
2. Integrated with Matplotlib and Pandas.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

# matplotlib

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
```

```python
plt.plot(var, var1)
plt.show()
```

# seaborn

```python
var2 = [1,2,3,4,5,6,7,8]
var3 = [8,7,6,5,4,3,2,1]
```

```python
sns.lineplot(x = var2, y = var3)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]

import pandas as pd

df = pd.DataFrame({"var":var, "var1":var1})

# df

sns.lineplot(x = 'var', y = 'var1', data = df)
plt.show()
```

# To give grid on graph

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=df1, hue='sex', palette='rocket_r', style='sex', markers=["o",">"])
plt.grid()
plt.show()
```

# To give Title on graph

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=df1, hue='sex', palette='rocket_r', style='sex', markers=["o",">"])
plt.grid()
plt.title("This is a graph related to penguins")
plt.show()
```
