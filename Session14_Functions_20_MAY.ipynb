```python
from datetime import datetime as dt

# Define current date
current_date = dt.today().date()

# Define another date for comparison
another_date = dt(2023, 5, 1).date()

# Year difference
year_difference = another_date.year - current_date.year

# Month difference
month_difference = (another_date.year - current_date.year) * 12 + (another_date.month - current_date.month)

# Day difference
day_difference = (another_date - current_date).days

print("Year Difference:", year_difference)
print("Month Difference:", month_difference)
print("Day Difference:", day_difference)
```
