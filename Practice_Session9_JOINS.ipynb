# MERGING TABLES
Before merging we need follow below instructions
1. identify the common fields in the datasets
2. define the left and the right table

```python
import pandas as pd
```

```python
file_path=r'C:\Users\ARPAN PANIGRAHI\Downloads\DATA ANALYST-DOCS\A3.HORIZONTAL JOIN DATA.xlsx'

stu_education=pd.read_excel(file_path,sheet_name='STU_EDUCATION')
stu_experience=pd.read_excel(file_path,sheet_name='STU_EXPERIENCE')

print(stu_education)
print() # For gap between two tables in output 
print(stu_experience)
```

```python
# Inner Join
# 1. LEFT TABLE
# 2. RIGHT TABLE
# 3. WHICH JOIN TO APPLY
# 4. COMMON KEY BETWEEN TWO TABLES
inner_join = pd.merge(left=stu_education,right=stu_experience,how='inner',on='STU_ID')
inner_join
```

```python
left_join = pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')
left_join
```

```python
right_join = pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')
right_join
```

```python
full_join = pd.merge(left=stu_education,right=stu_experience,how='outer',on='STU_ID')
full_join
```

```python
# LEFT NULL JOIN VALUES

left_null_join=pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')
left_null_join[left_null_join['COMPANY'].isnull()]
```

```python
# RIGHT NULL JOIN VALUES

right_null_join=pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')
right_null_join[right_null_join['STU_NAME'].isnull()]
```

```python
# UN-MATCHED JOIN

null_datasets=[left_null_join,right_null_join]

un_matched_join_values=pd.concat(null_datasets,ignore_index=True)
un_matched_join_values
```

# JOINING DATASETS WHERE BOTH TABLES HAVING SAME FIELD NAME

```python
import pandas as pd

SALES_2019 = pd.DataFrame({'PRODUCT':['APPLE','DELL','HP','ACER','LENOVO'],
                          'YEAR':[2019,2019,2019,2019,2019],
                          'SALES':[130,230,340,890,900]})
SALES_2019

SALES_2020 = pd.DataFrame({'PRODUCT':['APPLE','DELL','HP','ACER','LENOVO'],
                          'YEAR':[2020,2020,2020,2020,2020],
                          'SALES':[789,120,310,780,990]})
SALES_2020

print(SALES_2019)
print()
print(SALES_2020)
```

```python
sales_compare=pd.merge(left=SALES_2019,right=SALES_2020,how='inner',on='PRODUCT')
sales_compare
```

```python
sales_compare=pd.merge(left=SALES_2019,right=SALES_2020,how='inner',on='PRODUCT',suffixes=('_left','_right'))
sales_compare
```
