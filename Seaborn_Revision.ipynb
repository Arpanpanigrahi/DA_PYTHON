# Correlation with HeatMap

```python
1. Inter-relationship btw 2 classes.
2. Perform operation only on numerical values like integer and float. Ex- 1, 2.98
3. Range : +1 to -1.
```

```python
graph_v1.corr()
```

```python
# graph_v1 -- ['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']
# df --> graph_v1['total_bill', 'tip', 'size'] only numeric value in separate table

df = graph_v1[['total_bill', 'tip', 'size']]
df
# ['total_bill', 'tip', 'size'] -- List
```

```python
df.corr()
```

```python
# To know the range  --- Heatmap

sns.heatmap(df.corr())
```

```python
Three Types
1. Univariant     - sns.displot(x)
2. Bivariant      - sns.joinplot(x,y)
3. Multivariant   - sns.pairplot(data = table_name)
```

```python
# Univariant Analysis

sns.displot(df['tip'])
```

```python
sns.displot(df['tip'], kde=False)

# KDE - KERNEL DENSITY INFORMATION
```

```python
sns.displot(df['tip'], kde=False, bins=10)
```

```python
# Bivariant Analysis

sns.jointplot(x='tip',y='total_bill', data=df)
```

```python
# Bivariant Analysis

sns.jointplot(x='tip',y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, color='skyblue', marker='o', marginal_kws=dict(bins=15, fill=True))
```

```python
# Multivariant Analysis

sns.pairplot(df)
```
