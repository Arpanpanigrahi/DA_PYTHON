# MERGING TABLES IN PYTHON CODING

```python
import pandas as pd
import numpy as np

file_path=r'C:\Users\Debendra\BATCH_202207_PYTHON\SESSION-3\A2.RAW_DATA\A3.HORIZONTAL JOIN DATA-BASIC\A3.HORIZONTAL JOIN DATA.xlsx'

stu_education=pd.read_excel(file_path,sheet_name='STU_EDUCATION')
stu_experience=pd.read_excel(file_path,sheet_name='STU_EXPERIENCE')

print(stu_education)
print(stu_experience)
```

```python
# MERGING TABLES
# in this we have 2 tables
# 1. stu_education
# 2. stu_experience

# before merging we need follow below instructions
# 1. identify the common fields in the datasets
# 2. define the left and the right table
```

```python
# MERGING ARE 2 TYPES
# 1. INNER JOIN - IT TAKES ALL THE COMMON RECORDS
# 2. OUTER JOIN
#        FULL OUTER JOIN - IT TAKES BOTH COMMON AND UNCOMMON RECORDS FROM BOTH THE TABLES
#                UN-MATCHED JOIN - IT TAKES ONLT THE UN-COMMON RECORDS FROM BOTH THE TABLES
#        LEFT OUTER JOIN - IT TAKES ALL THE RECORDS FROM LEFT TABLE AND COMMON RECORDS FROM TEH RIGHT TABLE
#                LEFT NULL JOIN - IT TAKES REACORDS FROM THE LEFT TABLE WHICH IS NOT IN THE RIGHT TABLE
#        RIGHT OUTER JOIN - IT TAKES ALL THE RECORDS FROM THE RIGHT TABLE AND COMMON RECORDS FROM THE LEFT TABLE
#                RIGHT NULL JOIN - IT TAKES RECORDS IN THE RIGHT TABLE WHICH IS NOT IN THE LEFT TABLE
        
        
```

```python
# INNER JOIN
# 1. stu_education
# 2. stu_experience

# WHAT IS THE COMMON FIELD- STU_ID
# WHICH IS THE LEFT TABLE- stu_education
# WHICH IS THE RIGHT TABLE- Stu_experience

inner_join_values=pd.merge(left=stu_education,right=stu_experience,how='inner',on='STU_ID')

inner_join_values
```

```python
# FULL JOIN

full_join_values=pd.merge(left=stu_education,right=stu_experience,how='outer',on='STU_ID')

full_join_values
```

```python
# LEFT JOIN

left_join_values=pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')

left_join_values
```

```python
# RIGHT JOIN

right_join_values=pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')

right_join_values
```

```python
# NULL VALUES FOR NULL JOINS


# LEFT NULL JOIN VALUES

left_join_values=pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')
left_null_join_values=left_join_values[left_join_values['COMPANY'].isnull()]
left_null_join_values
```

```python
# RIGHT NULL JOIN VALUES

right_join_values=pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')
right_null_join_values=right_join_values[right_join_values['STU_NAME'].isnull()]
right_null_join_values
```

```python
# UN-MATCHED JOIN

null_datasets=[left_null_join_values,right_null_join_values]

un_matched_join_values=pd.concat(null_datasets,ignore_index=True)
un_matched_join_values
```

```python
import pandas as pd
import numpy as np

file_path=r'C:\Users\91824\OneDrive\Desktop\DV ANALYTICS\Assignments\PYTHON DEV SIR\SESSION-3-20220826T040134Z-001\SESSION-3\A2.RAW_DATA\A5.HORIZONTAL JOIN DATA-ADVANCE\A4.HORIZONTAL JOIN DATA.xlsx'

prod_sales=pd.read_excel(file_path,sheet_name='SUB_QUERY_DATA1')

prod_sales
```

```python
prod_sales_p1=prod_sales.groupby(['COUNTRY'],as_index=False).agg({'UNITS_SOLD':'sum'})
prod_sales_p1=prod_sales_p1.rename(columns={'UNITS_SOLD':'total_unit_sold'})
prod_sales_p1
```

```python
prod_sales_p2=pd.merge(left=prod_sales,right=prod_sales_p1,how='left',on='COUNTRY')
prod_sales_p2['ms']=(prod_sales_p2['UNITS_SOLD']/prod_sales_p2['total_unit_sold']).apply(lambda x: "{:.2%}".format(x))
prod_sales_p2
```

```python
# BEFORE APPROACHING TO A SUB-QUERIES WE MUST UNDERSTAND THE STEP BY APPROACHES

# step-1

prod_sales_summary=prod_sales.groupby(['COUNTRY'],as_index=False).agg({'UNITS_SOLD':'sum'})
prod_sales_summary=prod_sales_summary.rename(columns={'UNITS_SOLD':'TOTAL_SALES'})
prod_sales_summary
```

```python
# step-2

prod_sales_v1=pd.merge(left=prod_sales,right=prod_sales_summary,how='left',on='COUNTRY')
prod_sales_v1['MS']=(prod_sales_v1['UNITS_SOLD']/prod_sales_v1['TOTAL_SALES']).map(lambda n: '{:,.2%}'.format(n))
prod_sales_v1
```

```python
import pandas as pd
import numpy as np

file_path=r'C:\Users\Debendra\BATCH_202207_PYTHON\SESSION-3\A2.RAW_DATA\A7.DATA MODEL IN PYTHON\A1.DATA MODEL BY EXAMPLE DATA.xlsx'

transaction_master=pd.read_excel(file_path,sheet_name='TRANSACTION_TABLE')
customer_master=pd.read_excel(file_path,sheet_name='CUSTOMER_TABLE')
product_master=pd.read_excel(file_path,sheet_name='PRODUCT_TABLE')
store_master=pd.read_excel(file_path,sheet_name='STORE_TABLE')
staff_master=pd.read_excel(file_path,sheet_name='STAFF_MASTER')
supplier_master=pd.read_excel(file_path,sheet_name='SUPPLIER_MASTER')

print(transaction_master)
print(customer_master)
print(product_master)
print(store_master)
print(staff_master)
print(supplier_master)
```

```python
# 1. STAR SCHEMA
print(transaction_master)
print(customer_master)
print(product_master)
print(store_master)
print(staff_master)
```

```python
# print(transaction_master)
# print(customer_master)
# print(product_master)
# print(store_master)
# print(staff_master)

dataset1=pd.merge(left=transaction_master,right=customer_master,how='left',on='CUST_ID')
dataset2=pd.merge(left=dataset1,right=product_master,how='left',on='PROD_ID')
dataset3=pd.merge(left=dataset2,right=store_master,how='left',on='STORE_ID')
data_model1_output=pd.merge(left=dataset3,right=staff_master,how='left',on='STAFF_ID')


data_model1_output.to_excel(r'C:\Users\Debendra\BATCH_202207_PYTHON\SESSION-3\A4.OUTPUT\data_model1_output.xlsx',index=None)
print(data_model1_output)
```
