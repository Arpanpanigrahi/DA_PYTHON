```python
df1=df.sort_values(by=['STORE_NAME','CUSTOMER_ID','VISIT_DATE'],ascending=[True,True,True])
df1.head()
```

```python
df1['prev_spent']=df1.groupby(['STORE_NAME','CUSTOMER_ID'],as_index=False)['SPENT'].shift(1)
df1.fillna(0,inplace=True)
```

```python
df1['spent_diff']=abs(df1['SPENT']-df1['prev_spent'])
df1
```

```python
import datetime as dt
df1['prev_visit']=df1.groupby(['STORE_NAME','CUSTOMER_ID'])['VISIT_DATE'].shift(1)
df1['days_gape']=(df1['VISIT_DATE']-df1['prev_visit']).dt.days
df2=df1.groupby(['STORE_NAME','CUSTOMER_ID']).agg({'days_gape':'mean'})
df2
```
