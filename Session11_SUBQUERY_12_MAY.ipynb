```python
df_store_cust=df.groupby(['STORE_NAME','CUSTOMER_ID'],as_index=False).agg({'VISIT_DATE':'count'}).rename(columns={'VISIT_DATE':'total_visit'})
df_store_cust
```

```python
df_store_avg_trip=df_store_cust.groupby(['STORE_NAME'],as_index=False).agg({'total_visit':'mean'}).rename(columns={'total_visit':'str_wise_avg_visit'})
df_store_avg_trip
```

```python
import numpy as np
df2=pd.merge(left=df_store_cust,right=df_store_avg_trip,how='left',on=['STORE_NAME'])
# df2['vsit_segmentation']=np.where(df2['total_visit']<df2['str_wise_avg_visit'], 'NON_PREMIER','PREMIER')
df2['vsit_segmentation']='PREMIER'
df2['vsit_segmentation'][df2['total_visit']<=df2['str_wise_avg_visit']]='NON_PREMIER'
df2
```

```python
df2['add_10']=(df2['str_wise_avg_visit']/df2['total_visit']).apply(lambda x: "{:.2%}".format(x))
df2['add_10']=(df2['str_wise_avg_visit']/df2['total_visit']).apply(lambda x: "${:.2}".format(x))
df2
```
