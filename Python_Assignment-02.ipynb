```python
import pandas as pd

file_name = r"C:\Users\ARPAN PANIGRAHI\Downloads\DATA ANALYST-DOCS\TASK-4-12-MAY.xlsx"
df = pd.read_excel(file_name, sheet_name = "TASK-4")
df.head()
```

```python
df.columns
```

```python
# Q1.	GET STORE WISE CUSTOMERS AVG SPENT 

store_avg_spent = df.groupby('STORE_NAME')['SPENT'].mean().reset_index()
store_avg_spent.columns = ['STORE_NAME', 'AVG_AMOUNT_SPENT_PER_CUSTOMER']
store_avg_spent
```

```python
# Alternate Method
avg_spent_per_store = df.groupby('STORE_NAME')['SPENT'].mean().reset_index(name='AVG_SPENT')
avg_spent_per_store
```

```python
# Q2.	GET STORE WISE PROFITABLE CUSTOMERS BY
#1. NO. OF VISITS

# Group the data by 'STORE_NAME' and 'CUSTOMER_ID', and calculate total spent and number of visits
customer_summary = df.groupby(['STORE_NAME', 'CUSTOMER_ID']).agg({'SPENT': 'sum', 'VISIT_DATE': 'count'}).reset_index()
customer_summary
```

```python
# Calculate profitability (total spent / number of visits)
customer_summary['PROFITABILITY'] = customer_summary['SPENT'] / customer_summary['VISIT_DATE']
customer_summary
```

```python
# Sort customers by number of visits and profitability
profitable_customers = customer_summary.sort_values(by=['STORE_NAME', 'VISIT_DATE', 'PROFITABILITY'], ascending=[True, False, False])

# Print the result
profitable_customers
```

```python
# Q2.	GET STORE WISE PROFITABLE CUSTOMERS BY
#2. TOTAL SPENT

# Group the data by 'STORE_NAME' and 'CUSTOMER_ID', and calculate the total spent for each customer in each store
store_customer_spent = df.groupby(['STORE_NAME', 'CUSTOMER_ID'])['SPENT'].sum().reset_index()
store_customer_spent
```

```python
# Sort the results by total spent in descending order within each store
store_customer_spent_sorted = store_customer_spent.sort_values(by=['STORE_NAME', 'SPENT'], ascending=[True, False])
store_customer_spent_sorted
```

```python
# Get the most profitable customer for each store
most_profitable_customers = store_customer_spent_sorted.drop_duplicates(subset='STORE_NAME').reset_index()

# Print the result
most_profitable_customers
```

```python
#Q2. GET STORE WISE PROFITABLE CUSTOMERS BY
#3.	AVG SPENT 

# Group the data by 'STORE_NAME' and 'CUSTOMER_ID' and calculate the average spent for each customer in each store
store_customer_avg_spent = df.groupby(['STORE_NAME', 'CUSTOMER_ID'])['SPENT'].mean().reset_index()
store_customer_avg_spent
```

```python
# Rank customers within each store based on their average spent
store_customer_avg_spent['RANK'] = store_customer_avg_spent.groupby('STORE_NAME')['SPENT'].rank(ascending=False)
store_customer_avg_spent
```

```python
# Get the most profitable customer for each store
most_profitable_customers = store_customer_avg_spent[store_customer_avg_spent['RANK'] == 1]
most_profitable_customers
```

```python
# Sort the result by store name
most_profitable_customers = most_profitable_customers.sort_values(by='STORE_NAME')

# Print the result
most_profitable_customers
```

```python
# Q3. GET STORE WISE CUSTOMER VISITS_SEGMENTATION		
		
	# 	VISIT_SEGMENTATION
	# NO. OF VISITS >= AVG NO. OF VISITS THEN PREMIER
 #    NO. OF VISITS < AVG NO. OF VISITS ELSE	NON-PREMIER

df_store_cust=df.groupby(['STORE_NAME','CUSTOMER_ID'],as_index=False).agg({'VISIT_DATE':'count'}).rename(columns={'VISIT_DATE':'total_visit'})
df_store_cust
```

```python
df_store_avg_trip=df_store_cust.groupby(['STORE_NAME'],as_index=False).agg({'total_visit':'mean'}).rename(columns={'total_visit':'str_wise_avg_visit'})
df_store_avg_trip
```

```python
import numpy as np
df2=pd.merge(left=df_store_cust,right=df_store_avg_trip,how='left',on=['STORE_NAME'])
# df2['visit_segmentation']=np.where(df2['total_visit']<df2['str_wise_avg_visit'], 'NON_PREMIER','PREMIER')
df2['visit_segmentation']='PREMIER'
df2['visit_segmentation'][df2['total_visit']<=df2['str_wise_avg_visit']]='NON_PREMIER'
df2
```

```python
#Q4. GET STORE WISE CUSTOMER SPENT_SEGMENTATION		
		
	# 	SPENT_SEGMENTATION
	# SPENT  >= AVG SPENT	PREMIER
	# SPENT  < AVG SPENT	NON-PREMIER
df_store_cust = df.groupby(['STORE_NAME', 'CUSTOMER_ID'], as_index=False)['SPENT'].sum().rename(columns={'SPENT': 'total_spent'})
df_store_cust
```

```python
# Calculate the average spent for each store
df_store_avg_spent = df_store_cust.groupby('STORE_NAME', as_index=False)['total_spent'].mean().rename(columns={'total_spent': 'str_wise_avg_spent'})
df_store_avg_spent
```

```python
# Merge the average spent back to the customer dataframe
df3 = pd.merge(left=df_store_cust, right=df_store_avg_spent, how='left', on='STORE_NAME')

# Segment customers based on their spent compared to the store-wise average
df3['spent_segmentation'] = np.where(df3['total_spent'] >= df3['str_wise_avg_spent'], 'PREMIER', 'NON-PREMIER')

# Print the result
df3.head()
```

```python
# Q5. Final segmentation

# Merge the spent segmentation and visit segmentation dataframes
df_merged = pd.merge(left=df2, right=df3, how='left', on=['CUSTOMER_ID', 'STORE_NAME'])
df_merged
```

```python
#Define bucket conditions for customer segmentation
df_merged['Final_Segmenebation'] = 'BONAFIED'
df_merged['Final_Segmenebation'][(df_merged['visit_segmentation'] == 'PREMIER') & (df_merged['spent_segmentation'] == 'NON-PREMIER')]='POWER'
df_merged['Final_Segmenebation'][(df_merged['visit_segmentation'] == 'NON-PREMIER') & (df_merged['spent_segmentation'] == 'PREMIER')] = 'CASUAL'
df_merged['Final_Segmenebation'][(df_merged['visit_segmentation'] == 'NON-PREMIER') & (df_merged['spent_segmentation'] == 'NON-PREMIER')]= 'MISER'
df_merged


# SPENT_SEGMENTATION - df3
# VISIT_SEGMENTATION - df2
```
