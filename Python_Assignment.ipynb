```python
import pandas as pd

w = pd.read_csv(r"C:\Users\ARPAN PANIGRAHI\Downloads\DATA ANALYST-DOCS\MED_SELECT_V6.csv")
print(w)
```

```python
# Q1	CREATE A NEW TABLE MED_SELECT_V6 AND ADD THE FOLLOING COLUMNS AS	
# 	AGE_BUCKET	
# 		LESS THAN 30
# 		30 TO 50
# 		51 TO 70
# 		70+

w['Age_tag'] = 'Less_than_30'
w['Age_tag'][w['Age'].between(30,50)]='Less_than_50'
w['Age_tag'][w['Age'].between(51,70)]='Less_than_70'
w['Age_tag'][w['Age']>70]='More_than_70'
print(w)
```

```python
w
```

```python
# SPENT_BUCKET	
# 		LESS THAN 300
# 		301 TO 500
# 		501 TO 700
# 		701 TO 1000
# 		1000+


w['Spent_tag'] = 'Less_than_300'
w['Spent_tag'][w['Spent amount'].between(300,500)]='Less_than_500'
w['Spent_tag'][w['Spent amount'].between(501,700)]='Less_than_700'
w['Spent_tag'][w['Spent amount'].between(701,1000)]='Less_than_1000'
w['Spent_tag'][w['Spent amount']>1000]='More_than_1000+'
print(w)
	
```

```python
w.columns
```

```python
# CUSTOMER_SEGMENT	
# 	IF AGE BETWEEN 30 AND 70 AND GENDER=FEMALE AND SPENT_AMOUNT >=300	PREMIER
# 	IF AGE BETWEEN 30 AND 70 AND GENDER=MALE AND SPENT_AMOUNT >=300	PREMIER
# 	ELSE 	NON-PREMIER

w['Age_tag'] = 'NON-PREMIER'
w['Age_tag'][(w['Age'].between(30,70)) & (w['GENDER']=='Female') & (w['Spent amount'] >= 300)] = 'PREMIER'
w['Age_tag'][(w['Age'].between(30,70)) & (w['GENDER']=='Male') & (w['Spent amount'] >= 300)] = 'PREMIER'

w
```
