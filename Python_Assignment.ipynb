```python
# Q1	FROM CUSTOMER_MONTHLY_SALES TABLE GET
	# Q1+Q2+Q3+Q4 AS TOTAL SALES

df1 = pd.read_csv(r"C:\Users\ARPAN PANIGRAHI\Downloads\DATA ANALYST-DOCS\CUSTOMER_MONTHLY_SALES.csv")
print(df1)
```

```python
# Q1 FROM CUSTOMER_MONTHLY_SALES TABLE GET Q1+Q2+Q3+Q4 AS TOTAL SALES

df1['Q1_Total_Sales'] = df1['JAN']+df1['FEB']+df1['MAR']
df1['Q2_Total_Sales'] = df1['APR']+df1['MAY']+df1['JUN']
df1['Q3_Total_Sales'] = df1['JUL']+df1['AUG']+df1['SEP']
df1['Q4_Total_Sales'] = df1['OCT']+df1['NOV']+df1['DEC']
df1.head()
```

```python
df1['Total_Yearly_Sales'] = df1['Q1_Total_Sales'] + df1['Q2_Total_Sales'] + df1['Q3_Total_Sales'] + df1['Q4_Total_Sales']
df1.head()
```

```python
# Q2	GET SALES IN %
# 	Q1 SALES % OVER YEARLY SALES
# 	Q2 SALES % OVER YEARLY SALES
# 	Q3 SALES % OVER YEARLY SALES
# 	Q4 SALES % OVER YEARLY SALES


df1['Sales_Percentage'] = (df1['Q1_Total_Sales'] + df1['Q2_Total_Sales'] + df1['Q3_Total_Sales'] + df1['Q4_Total_Sales'] / df1['Total_Yearly_Sales']) * 100
df1
```
